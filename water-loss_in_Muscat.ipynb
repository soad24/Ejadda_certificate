```python
# STEP 1: Install Prophet (this takes ~1 min)
!pip install prophet

# STEP 2: Import libraries
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# STEP 3: Upload your file
from google.colab import files
uploaded = files.upload()

# STEP 4: Load your CSV (muscat_loss.csv)
df = pd.read_csv("muscat_loss.csv")  # replace with your file name
df = df.rename(columns={"Date": "ds", "Calculated_Loss_Percent": "y"})
df['ds'] = pd.to_datetime(df['ds'])
df = df.dropna(subset=['ds', 'y'])

# STEP 5: Fit Prophet model
model = Prophet()
model.fit(df)

# STEP 6: Forecast next 6 months
future = model.make_future_dataframe(periods=6, freq='M')
forecast = model.predict(future)

# STEP 7: Plot forecast
fig1 = model.plot(forecast)
plt.title("Water Loss % Forecast - Muscat")
plt.grid(True)
plt.show()

# STEP 8: Show forecast table
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(6)
```
